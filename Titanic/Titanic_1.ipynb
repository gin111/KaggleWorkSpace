```python
import pandas as pd
import csv as csv
from sklearn.ensemble import RandomForestClassifier
```

```python
# Load training data
train_df = pd.read_csv("train.csv", header=0)
train_df.head(10)
```

```python
# Convert "Sex" to be a dummy variable (female = 0, Male = 1)
train_df["Gender"] = train_df["Sex"].map({"female": 0, "male": 1}).astype(int)
train_df.head(10)
```

```python
train_df.Age[train_df.Age.isnull()]
```

```python
median_age = train_df["Age"].dropna().median()
median_age
```

```python
# Complement the missing values of "Age" column with average of "Age"
if len(train_df.Age[train_df.Age.isnull()]) > 0:
    train_df.loc[(train_df.Age.isnull()), "Age"] = median_age
median_fare = train_df["Fare"].dropna().median()
if len(train_df.Fare[train_df.Fare.isnull()]) > 0:
    train_df.loc[(train_df.Fare.isnull()), "Fare"] = median_fare
train_df.head(10)
```

```python
# remove un-used columns
train_df = train_df.drop(["Name", "Ticket", "Sex", "SibSp", "Parch", "Cabin", "Embarked", "PassengerId"], axis=1)
train_df.head()
```

```python
train_df.values
```

```python
train_data = train_df.values
model = RandomForestClassifier(n_estimators=100)
output = model.fit(train_data[0::, 1::], train_data[0::, 0])
```

```python
accuracy = model.score(train_data[0::, 1::], train_df.Survived)
print('accuracy {0:.2%}'.format(accuracy))
```

```python
# Load test data, Convert "Sex" to be a dummy variable
test_df = pd.read_csv("test.csv", header=0)
test_df["Gender"] = test_df["Sex"].map({"female": 0, "male": 1}).astype(int)

# Complement the missing values of "Age" column with average of "Age"
median_age = test_df["Age"].dropna().median()
if len(test_df.Age[test_df.Age.isnull()]) > 0:
     test_df.loc[(test_df.Age.isnull()), "Age"] = median_age
median_fare = test_df["Fare"].dropna().median()
if len(test_df.Fare[test_df.Fare.isnull()]) > 0:
    test_df.loc[(test_df.Fare.isnull()), "Fare"] = median_fare        
        
test_df.head(10)
```
